```python
import os
import shutil
from pathlib import Path

# Configuration
SOURCE_BASE = Path("data/fruits-360")
TARGET_BASE = Path("PrepData")
SPLITS = ["Training", "Validation", "Test"]

# Define your mapping; we use not all the data 
# Example: Class_0 = Apples, Class_1 = Bananas
MAPPING = {
    "Class_0": ["Apple 10"], 
    "Class_1": ["Papaya 2"]
}

def prepare_binary_dataset(mapping, source_base, target_base):
    if target_base.exists():
        print(f"Removing existing directory: {target_base}")
        shutil.rmtree(target_base)
    
    for target_class, source_categories in mapping.items():
        for split in SPLITS:
            dest_dir = target_base / split / target_class
            dest_dir.mkdir(parents=True, exist_ok=True)
            
            counter = 0
            for cat in source_categories:
                src_dir = source_base / split / cat
                
                if not src_dir.exists():
                    print(f"Warning: {src_dir} not found. Skipping.")
                    continue
                
                for img_path in src_dir.glob("*"):
                    if img_path.suffix.lower() in [".jpg", ".jpeg", ".png"]:
                        # Use symlinks to save disk space
                        os.symlink(img_path.absolute(), dest_dir / f"{cat}_{img_path.name}")
                        counter += 1
            
            print(f"Created {split}/{target_class}: {counter} images.")

# Execute
prepare_binary_dataset(MAPPING, SOURCE_BASE, TARGET_BASE)
```
